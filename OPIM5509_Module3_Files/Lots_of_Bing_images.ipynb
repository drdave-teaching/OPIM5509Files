```python
!pip install bing-image-downloader

from bing_image_downloader import downloader

# Define search queries
queries = ['Cars', 'Motorcycles']
# Additional filters or criteria to refine the search within the main categories defined in the queries list.
subqueries = ['White', 'Black', 'Blue', 'Red', 'yellow', 'Green', 'Orange', 'White and Black', 'Orange and Green', 'Green and Black']

# Download images for
for query in queries:
  for subquery in subqueries:
    full_query = f'{query} {subquery}'
    downloader.download(full_query, limit=100, output_dir='images', adult_filter_off=True, force_replace=False, timeout=300)

# Create images_dataset : folders based on categories
import os

destination_folder = 'images_dataset'

if not os.path.exists(destination_folder):
  os.makedirs(destination_folder)

# Create directories for ['Cars', 'Motorcycles']
for category in ['Cars', 'Motorcycles']:
    for query in queries:
      if not os.path.exists(f'{destination_folder}/{category}'):
        os.makedirs(f'{destination_folder}/{category}', exist_ok=True)

# This code will copy all files from images to images_dataset based in the type of images
import os
import shutil

images_path = '/content/images' # images directory
imageFolders = [folder for folder in os.listdir(images_path) if os.path.isdir(os.path.join(images_path, folder))]

images_dataset_path = '/content/images_dataset'  # images_dataset directory
images_dataset_Folders = [folder for folder in os.listdir(images_dataset_path) if os.path.isdir(os.path.join(images_dataset_path, folder))]

# This code will copy the all files from images to respective images_dataset
for images_dataset_Folder in images_dataset_Folders:
  for imagesFolder in imageFolders:
    if imagesFolder.startswith(images_dataset_Folder):
      source_folder = os.path.join(images_path, imagesFolder)
      destination_folder = os.path.join(images_dataset_path, images_dataset_Folder)
      if os.path.exists(source_folder):
       for file_name in os.listdir(source_folder):
        source_file_path = os.path.join(source_folder, file_name)
        destination_file_path = os.path.join(destination_folder, file_name)
        # Check if the file is an image before copying (only jpg extn)
        if file_name.lower().endswith(('.jpg')):
         # Check if the file already exists in the destination folder
         if os.path.exists(destination_file_path):
            # Create a unique file name by appending a number
            base_name, ext = os.path.splitext(file_name)
            counter = 1
            while os.path.exists(destination_file_path):
                new_file_name = f"{base_name}_{counter}{ext}"
                destination_file_path = os.path.join(destination_folder, new_file_name)
                counter += 1
         shutil.copy(source_file_path, destination_file_path)

# Delete the source folder after copying its contents
source_folder_path = '/content/images'
shutil.rmtree(source_folder_path)
print(f'Deleted folder: {source_folder_path}')

# Display number of Car Images
!ls -1 /'content'/'images_dataset'/'Cars' | wc -l

# Display number of Motorcycles Images
!ls -1 /'content'/'images_dataset'/'Motorcycles' | wc -l

# Use the 'images_dataset' for partitions and subsequent steps.
```
